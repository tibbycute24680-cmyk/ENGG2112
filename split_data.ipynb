```python
#code to split data from large, combined csv file to training, testing, 
#and validation data by airport. 

#date is in YYYY-MM-DD format

import pandas as pd
from pathlib import Path

DATA_DIR = Path.home() 

FILES = [
    "2025_merged_dataset2.csv",
    "2023_merged_dataset1.csv"
]

#32 airpors for training
training = [
    "ATL", "ORD", "LAX", "DFW", "DEN", "JFK", "CLT", "SEA", "MIA", "PHX",
    "IAH", "BOS", "MSP", "SFO", "EWR", "LGA", "DTW", "PHL", "FLL", "SLC",
    "SAN", "BWI", "MDW", "LAS", "MCO", "RDU", "MSY", "BNA", "STL", "TPA",
    "AUS", "HNL"
]

#9 airports for validation
validation = [
    "SJC", "OAK", "SMF",
    "LGA", "DTW", "PHL",
    "FLL", "MDW", "LAS"
]

#9 airports for testing
test = [
    "PHX", "IAH", "BOS",
    "SLC", "SAN", "BWI",
    "SAT", "PIT", "CMH"
]


```

```python
#split and write to new csv files

for file in FILES:
    df = pd.read_csv(DATA_DIR / file)

    train_df = df[df["airport"].isin(training)]
    val_df   = df[df["airport"].isin(validation)]
    test_df  = df[df["airport"].isin(test)]

    base_name = file.replace(".csv", "")

    train_df.to_csv(DATA_DIR / f"{base_name}_train.csv", index=False)
    val_df.to_csv(DATA_DIR / f"{base_name}_val.csv", index=False)
    test_df.to_csv(DATA_DIR / f"{base_name}_test.csv", index=False)

print("Done!")
```

```python
#debugging, ensure split worked and all have correct number of rows
splits = ["train", "val", "test"]

for file in FILES:
    base = file.replace(".csv", "") 
    print(f"\n=== {file} ===")
    
    for split in splits:
        path = DATA_DIR / f"{base}_{split}.csv"
        df = pd.read_csv(path)
        print(f"{split}: {len(df)} rows")
```
